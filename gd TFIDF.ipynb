```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
final_corpus = data.drop_duplicates(subset='tweet', keep='first', inplace=False)
```

```python
final_corpus.shape
```

```python
stopwords= set(['br', 'the', 'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've",\
            "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', \
            'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their',\
            'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', \
            'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', \
            'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', \
            'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after',\
            'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further',\
            'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more',\
            'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', \
            's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', \
            've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn',\
            "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn',\
            "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", \
            'won', "won't", 'wouldn', "wouldn't"])
```

```python
import re

def cleaning(series):
    processed_reviews = []
    for text in series.values:
        text = re.sub('[^A-Za-z0-9]+', ' ', text)
        text = re.sub("\S*\d\S*", "", text).strip()
        text = ' '.join(e.lower() for e in text.split() if e.lower() not in stopwords)
        processed_reviews.append(text)
    return processed_reviews
processed_tweets = cleaning(final_corpus['tweet'])
processed_tweets[0:2]
```

```python
# updated stopwords
for _ in range(0,len(processed_tweets)):
    for word in processed_tweets[_].split():
        if (len(word)<2) and (word.lower() not in stopwords):
            stopwords.add(word.lower())

print(stopwords)
```

```python
processed_tweets = cleaning(pd.Series(processed_tweets))
processed_tweets[:5]
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
stemmed_text = list()
for i in range(len(processed_tweets)):
    stemmed_text.append(' '.join(porter.stem(word) for word in processed_tweets[i].split()))
stemmed_text[:5]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vect = TfidfVectorizer(ngram_range=(1,2), min_df=5)
tf_idf_vect.fit(stemmed_text)
len(tf_idf_vect.get_feature_names())
```

```python
final__tf_idf = tf_idf_vect.transform(stemmed_text)
print(type(final__tf_idf))
print(final__tf_idf.get_shape())
```

```python
X = pd.DataFrame(final__tf_idf.toarray())
y = final_corpus['label']
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.metrics import f1_score
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X, y)
print('train accuracy score : ', lr.score(X_train, y_train))
# print('test accuracy score : ', lr.score(X_test, y_test))
print('train f1 score : ', f1_score(y_train, lr.predict(X_train)))
# print('test f1 score : ', f1_score(y_test, lr.predict(X_test)))
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X, y)
print('train accuracy score : ', nb.score(X_train, y_train))
# print('test accuracy score : ', nb.score(X_test, y_test))
print('train f1 score : ', f1_score(y_train, nb.predict(X_train)))
# print('test f1 score : ', f1_score(y_test, nb.predict(X_test)))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('train accuracy score : ', knn.score(X_train, y_train))
print('test accuracy score : ', knn.score(X_test, y_test))
print('train f1 score : ', f1_score(y_train, knn.predict(X_train)))
print('test f1 score : ', f1_score(y_test, knn.predict(X_test)))
```

## SVM

```python
from sklearn.svm import SVC

sm = SVC(kernel='linear',class_weight='balanced')
sm.fit(X, y)
print('train accuracy score : ', sm.score(X_train, y_train))
# print('test accuracy score : ', sm.score(X_test, y_test))
print('train f1 score : ', f1_score(y_train, sm.predict(X_train)))
# print('test f1 score : ', f1_score(y_test, sm.predict(X_test)))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=30, class_weight='balanced')
rf.fit(X_train, y_train)
print('train accuracy score : ', rf.score(X_train, y_train))
print('test accuracy score : ', rf.score(X_test, y_test))
print('train f1 score : ', f1_score(y_train, rf.predict(X_train)))
print('test f1 score : ', f1_score(y_test, rf.predict(X_test)))
```
